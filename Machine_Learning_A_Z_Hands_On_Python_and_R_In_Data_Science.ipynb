###One-hot encoding


One-hot encoding is typically used for categorical variables or features in your dataset. Categorical variables represent qualitative data that can take on a limited number of distinct categories or levels.

Here are some examples of categorical variables that are commonly one-hot encoded:

* Color: Red, Blue, Green
* Size: Small, Medium, Large
* Country: USA, Canada, UK, Australia
* Gender: Male, Female
* Payment Method: Credit Card, Debit Card, PayPal

These variables have discrete categories that do not have any inherent order or numerical meaning. One-hot encoding is used to represent each category as a binary feature, where each category is assigned a separate column with 0s and 1s indicating the presence or absence of that category.

Numeric variables or continuous variables, such as age or income, do not require one-hot encoding as they already represent numerical values.

It's important to note that the decision to perform one-hot encoding depends on the specific requirements of your analysis or machine learning model.

```python
# Perform one-hot encoding
X_df_encoded = pd.get_dummies(X_df['Country'])

# Concatenate the encoded columns with the original dataframe
X_df_encoded = pd.concat([X_df, X_df_encoded], axis=1)
#X_df_encoded = X_df_encoded.drop('Country', axis=1)
X_df_encoded.head()
```

###Label Encoding

Label Encoding is used for handling ordinal data:

* Education Levels: Converting categories like : ["High School", "Bachelor's", "Master's", "PhD"] to [0, 1, 2, 3].
* Customer Satisfaction: Converting responses such as : ["Very Unsatisfied", "Unsatisfied", "Neutral", "Satisfied", "Very Satisfied"] to [0, 1, 2, 3, 4].
* target variable or dependent variable
