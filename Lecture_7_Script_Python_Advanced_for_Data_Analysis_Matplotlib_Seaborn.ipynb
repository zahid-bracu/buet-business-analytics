##Line Chart

####Only one Y axis Values

```python
import matplotlib.pyplot as plt

# Sample data
y = [2, 3, 5, 7, 11, 13]

# Create the figure and plot
plt.figure(figsize=(6, 4))

# Plot the first line
plt.plot(y,
         label='Increment',
         marker='o',
         markersize=8,
         markeredgecolor='blue',
         markeredgewidth=2,
         markerfacecolor='red',
         linestyle='--',
         linewidth=2,
         alpha=1)

# Adding titles and labels
plt.title('Increment Diagram')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')

# Adding a legend
plt.legend(title='Legend')

# Show the plot
plt.grid(True)
plt.show()
```

* label='Increment': Sets the label for the line, which will appear in the legend.
* marker='o': Specifies that circular markers should be used.
* markersize=8: Sets the size of the markers.
* markeredgecolor='blue': Sets the color of the marker edges to blue.
* markeredgewidth=2: Sets the width of the marker edges.
* markerfacecolor='red': Sets the color of the marker faces to red.
* linestyle='--': Sets the line style to dashed.
* linewidth=2: Sets the width of the line.
* alpha=1: Sets the opacity of the line to fully opaque.


####One X Axis & one Y axis

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5, 6]
y = [2, 3, 5, 7, 11, 13]

plt.figure(figsize=(6, 4)) # Create the figure and plot

plt.plot(x, y,  # Plot the first line
         label='Increment',
         marker='s',
         markersize=8,
         markeredgecolor='yellow',
         markeredgewidth=2,
         markerfacecolor='red',
         linestyle='--',
         linewidth=2,
         alpha=1)

plt.title('Increment By month') # Adding titles and labels
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')

plt.legend(title='Legend') # Adding a legend

plt.grid(True) # Show the plot
plt.show()
```

####One X axis & Two Y Axis

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5, 6]
y1 = [2, 3, 5, 7, 11, 13]
y2 = [1, 4, 6, 8, 10, 12]

plt.figure(figsize=(6, 4)) # Create the figure and plot

plt.plot(x, y1, # Plot the first line
         label='Prime Numbers',
         marker='o',
         markersize=8,
         markeredgecolor='blue',
         markeredgewidth=2,
         markerfacecolor='white',
         linestyle='-',
         linewidth=2,
         alpha=0.8)

plt.plot(x, y2, # Plot the second line
         label='Even Numbers',
         marker='s',
         markersize=8,
         markeredgecolor='red',
         markeredgewidth=2,
         markerfacecolor='yellow',
         linestyle='--',
         linewidth=2,
         alpha=0.8)

# Adding titles and labels
plt.title('Prime Numbers vs Even Numbers')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')

plt.legend(title='Legend') # Adding a legend

plt.grid(True)
plt.show() # Show the plot
```
