##Line Chart

####Only one Y axis Values

```python
import matplotlib.pyplot as plt

# Sample data
y = [2, 3, 5, 7, 11, 13]

# Create the figure and plot
plt.figure(figsize=(6, 4))

# Plot the first line
plt.plot(y,
         label='Increment',
         marker='o',
         markersize=8,
         markeredgecolor='blue',
         markeredgewidth=2,
         markerfacecolor='red',
         linestyle='--',
         linewidth=2,
         alpha=1)

# Adding titles and labels
plt.title('Increment Diagram')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')

# Adding a legend
plt.legend(title='Legend')

# Show the plot
plt.grid(True)
plt.show()
```

* label='Increment': Sets the label for the line, which will appear in the legend.
* marker='o': Specifies that circular markers should be used.
* markersize=8: Sets the size of the markers.
* markeredgecolor='blue': Sets the color of the marker edges to blue.
* markeredgewidth=2: Sets the width of the marker edges.
* markerfacecolor='red': Sets the color of the marker faces to red.
* linestyle='--': Sets the line style to dashed.
* linewidth=2: Sets the width of the line.
* alpha=1: Sets the opacity of the line to fully opaque.

####One X axis & Two Y Axis

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5, 6]
y1 = [2, 3, 5, 7, 11, 13]
y2 = [1, 4, 6, 8, 10, 12]

plt.figure(figsize=(6, 4)) # Create the figure and plot

plt.plot(x, y1, # Plot the first line
         label='Prime Numbers',
         marker='o',
         markersize=8,
         markeredgecolor='blue',
         markeredgewidth=2,
         markerfacecolor='white',
         linestyle='-',
         linewidth=2,
         alpha=0.8)

plt.plot(x, y2, # Plot the second line
         label='Even Numbers',
         marker='s',
         markersize=8,
         markeredgecolor='red',
         markeredgewidth=2,
         markerfacecolor='yellow',
         linestyle='--',
         linewidth=2,
         alpha=0.8)

# Adding titles and labels
plt.title('Prime Numbers vs Even Numbers')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')

plt.legend(title='Legend') # Adding a legend

plt.grid(True)
plt.show() # Show the plot
```

####Scattered Plot With Seaborn

A scatter plot is a type of data visualization that displays values for typically two variables for a set of data. The data is displayed as a collection of points, each having the value of one variable determining the position on the horizontal axis (x-axis) and the value of the other variable determining the position on the vertical axis (y-axis). Scatter plots are useful for showing relationships between two variables and can reveal correlations, trends, clusters, and outliers in the data.
Key Features of Scatter Plots

Axes:
* X-axis (Horizontal Axis): Represents one of the variables.
* Y-axis (Vertical Axis): Represents the other variable.

Data Points:
* Each point on the scatter plot represents an individual data point in the dataset.

Variables:
* Typically, scatter plots are used to compare two numerical variables, but additional dimensions can be represented using colors, sizes, or shapes.

Color/Hue:
* Scatter plots can include a categorical variable by using different colors for different categories, which can help in distinguishing groups within the data.


```python
import numpy as np
import seaborn as sns

# Generate random data
np.random.seed(42)
x = np.random.rand(100) * 100  # Random numbers between 0 and 100
y = x + (np.random.rand(100) * 50)  # Add some noise
category = np.random.choice(['A', 'B'], size=100)  # Random categories

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y, 'category': category})

# Set the figure size
plt.figure(figsize=(6, 4))

# Scatter plot with seaborn
sns.scatterplot(data=data, x='x', y='y', hue='category',s=100)

plt.title('Scatter Plot of Sample Data')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.show()
```

Description of the Example Scatter Plot

Axes:
* The X-axis represents the x values, ranging from 0 to 100.
* The Y-axis represents the y values, which are a noisy linear function of x.

Data Points:
* Each point represents a pair of x and y values from the dataset.

Color/Hue:
* Points are colored based on the category variable (A or B), which helps in distinguishing between the two categories.

Title and Labels:
* The plot has a title ("Scatter Plot of Sample Data") and labeled axes ("X Values" and "Y Values") to provide context.

####Scattered Plot with Matplot lib

```python
import numpy as np
import seaborn as sns

# Generate random data
np.random.seed(42)
x = np.random.rand(100) * 100  # Random numbers between 0 and 100
y = x + (np.random.rand(100) * 50)  # Add some noise
category = np.random.choice(['A', 'B'], size=100)  # Random categories
data = pd.DataFrame({'x': x, 'y': y, 'category': category}) # Create a DataFrame
plt.figure(figsize=(6, 4)) # Set the figure size

# Scatter plot with matplotlib
for category in data['category'].unique():
    subset = data[data['category'] == category]
    plt.scatter(subset['x'], subset['y'], label=category)
plt.title('Scatter Plot of Sample Data')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.legend()
plt.show()
```

Explanation :

Iterating Over Unique Categories:
* for category in data['category'].unique(): This loop iterates over each unique value in the category column of the DataFrame data. The unique() method returns an array of unique category values.

Creating Subsets of Data:
* subset = data[data['category'] == category]: For each category, a subset of the DataFrame data is created. This subset contains only the rows where the category column matches the current category in the loop.

Plotting the Data Points:
* plt.scatter(subset['x'], subset['y'], label=category): This creates a scatter plot for the current subset of data. The x values come from the x column, and the y values come from the y column. The label parameter is used to label the points by category, which will be used in the legend.
